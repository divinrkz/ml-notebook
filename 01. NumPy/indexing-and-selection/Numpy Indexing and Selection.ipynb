# Numpy Indexing and Selection

```python
import numpy as np
```

```python
arr = np.arange(0, 11)
```

```python
arr
```

```python
arr[0:4]
```

```python
arr[3:4]
```

```python
arr[:4]
```

```python
arr[4:]
```

```python
arr[:]
```

```python
arr[:1] = 100
```

```python
arr
```

```python
slice_of_arr = arr[1:3]
```

```python
slice_of_arr
```

```python
slice_of_arr[:] = 100
```

```python
slice_of_arr
```

```python
arr
```

```python
arr_copy = arr.copy()
```

```python
arr_copy
```

```python
arr_copy[:3] = np.arange(0, 3)
```

```python
arr_copy
```

```python
arr_2d = np.array([[5, 10, 15], [20, 25, 30], [35, 40, 45]])
```

```python
arr_2d
```

```python
arr_2d[1][2]
```

```python
arr_2d[1, 2]
```

```python
arr_2d[1:2, :]
```

```python
arr = np.arange(11)
```

```python
arr[arr > 3]
```

```python
arr_2d = np.arange(50).reshape(5, 10)
```

```python
arr_2d
```

```python
arr_2d[1:3, 3:5]
```
